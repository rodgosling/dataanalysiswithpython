# Series

### Series Creation

```python
import numpy as np
import pandas as pd
```

```python
# Given the X python list convert it to an Y pandas Series

flwrs = [10.8, 2.3, 2.1, 2.1, 1.8, 1.8, 1.4, 1.3, 1.0, 1.0, 0.983, 0.616, 0.512, 0.511, 0.502, 0.496, 0.245, 0.176]

dragseries = pd.Series (flwrs)
dragseries
```

```python
# Given the X pandas Series, show its values

dragseries.values
```

# Data Frame

## DataFrame creation

```python
# Create an empty pandas DataFrame

pd.DataFrame (data = [None], 
              index = [None],
             columns = [None])
```

```python
# Create a drag_df pandas DataFrame with the given data
drags = [
            ['Pabllo Vittar', 10.8, 'BR'],
            ['Bianca Del Rio', 2.3, 'US'],
            ['Adore Delano', 2.1, 'US'],
            ['Katya Zamo', 2.1, 'US'],
            ['Gloria Groove', 1.8, 'BR'],
            ['Alyssa Edwards', 1.8, 'US'],
            ['Shangela', 1.4, 'US'],
            ['Shea Couleé', 1.3, 'US'],
            ['Monét X Change', 1.0, 'US'],
            ['Yvie Oddly', 1.0, 'US'],
            ['Latrice Royale', 0.983, 'US'],
            ['Lorelay Fox', 0.616, 'BR'],
            ['BeBe Zahara Benet', 0.512, 'US'],
            ['Aretuza Lovi', 0.511, 'BR'],
            ['Rita von Hunty', 0.502, 'BR'],
            ['Lia Clark', 0.496, 'BR'],
            ['Silvetty Montilla', 0.245, 'BR'],
            ['Coco Montrese', 0.176, 'US']
]

drags_df = pd.DataFrame (data = drags)
drags_df
```

```python
columns_names = ['Nome', 'Followers', 'País']

drags_df.columns = columns_names
drags_df
```

```python
# Add index names to the marvel_df (use the name as index)

drags_df.index = drags_df ['Nome']
drags_df
```

```python
# Drop the name column as it's now the index

drags_df = drags_df.drop ('Nome', axis = 1)
drags_df

# axis = 0 -- rows
# axis = 1 -- columns
```

<hr style="margin-bottom: 40px;">

## DataFrame manipulation and operations

```python
# Modify the country of 'Coco Montrese' to United States

drags_df.loc ['Coco Montrese', 'País'] = 'United States'
drags_df
```

```python
# Drop row where index is Lia Clark

drags_df.drop (index = 'Lia Clark')

# para deletar definitivamente: drags_df.drop (index = 'Lia Clark', inplace = True)
```

```python
# Replace US to United States

drags_df.replace ('US','United States', inplace = True)
drags_df
```

```python
# Replace BR to Brasil

drags_df.replace ('BR','Brasil', inplace = True)
drags_df
```

```python
# Add a column Idioma based on column País. If País = Brasil, Idioma =  Português / If País = United States, Idioma =  Inglês

drags_df ['Idioma'] = np.where(drags_df['País'].str.contains('Brasil'),'Português','Inglês')
drags_df
```

<hr style="margin-bottom: 40px;">

## DataFrame boolean arrays (also called masks)

```python
# Given the drags_df pandas DataFrame, make a mask showing the portuguese speakers

mask = drags_df['Idioma']=='Português'
drags_df[mask]
```

```python
# Given the drags_df pandas DataFrame, get the drags with followers >= 1.0 million

drags_df [drags_df['Followers'] >= 1.0]
```

```python
# Given the drags_df pandas DataFrame, get the drags with followers >= 1.0 million who speak portuguese

drags_df [(drags_df['Followers'] >= 1.0) & (drags_df['Idioma'] == 'Português')]
```
